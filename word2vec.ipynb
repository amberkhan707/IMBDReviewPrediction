## Word to vector Conversion:

```python
from nltk.tokenize import sent_tokenize
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
```

```python
sentence = """MD AMBER KHAN. I am currently learning ML and GenAI. I am trying to be good in GenAI and make a Production level working project.
I will definately complete this easily"""
```

```python
sentence = sent_tokenize(sentence)
sentence
```

```python
# First i will change this into one_hot
vocab = 10000
dimension = 28
max_sent_length = 20
```

```python
onehot = [one_hot(words,vocab)for words in sentence]
```

```python
embedded_doc = pad_sequences(onehot)
embedded_doc
```

```python
model = Sequential()
model.add(Embedding(vocab, dimension, input_length = max_sent_length))
model.compile('adam', 'mse')
```

```python
model.predict(embedded_doc)
```
