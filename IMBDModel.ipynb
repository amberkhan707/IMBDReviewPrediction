```python
from tensorflow.keras.datasets import imdb
import numpy as np
data = imdb.load_data(num_words=10000)
print(data) # X_train data, y_train data and X_test data, y_test data is already splitted
(X_train,y_train), (X_test,y_test) = data 
print(X_train[0])
x = np.max(X_train)
print(x)
```

```python
# Converting this numerical into words 
words = imdb.get_word_index()
word_to_index = {value:key for key,value in words.items()}
first_sent = X_train[0]
sentence = ' '.join([word_to_index.get(num -3,'!' ) for num in first_sent])
print(sentence)
```

```python
#Applied Padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
X_train = pad_sequences(X_train, 500)
X_test = pad_sequences(X_test, 500)
```

```python
# We are going to train the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, SimpleRNN

model = Sequential()
model.add(Embedding(input_dim = 10000, output_dim = 128, input_length = 500))
model.add(SimpleRNN(64,activation ='relu'))
model.add(Dense(1,activation ='sigmoid'))
model.compile(optimizer='adam', loss= 'binary_crossentropy', metrics= ['accuracy'])
model.build(input_shape=(None,500))
model.summary()
```

```python
# Applied Early Stopping
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(patience = 5, restore_best_weights = True, monitor = 'val_loss')
early_stop
```

```python
# Fitted the model with our data
model.fit(X_train, y_train, validation_data=(X_test,y_test), batch_size=32, epochs=10, callbacks= [early_stop])
```

```python
# Model is saved to used in another files
model.save('SimpleRNNmodel.h5')
```
